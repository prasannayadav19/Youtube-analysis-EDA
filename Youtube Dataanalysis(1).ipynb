```python
#first select genre then select the channel and then user will get the videos:
#here we created a dropdown for genre and called in a function
k1 = widgets.Dropdown(
    options=video['Genre'].unique(),
    value='Music',
    description='Genre:',
    disabled=False,
)
def genre(m):
    gen=video[video['Genre']==m]
    k2 = widgets.Dropdown(
        options=video['channelTitle'].unique(),
        value=video['channelTitle'].unique()[0],
        description='channel:',
        disabled=False,
    )
    def channel(d):
      y=gen.sort_values(by = 'view_count',ascending=False).head()
      sns.barplot(y = 'title',x = 'view_count',data=y)
      plt.title(d)
      plt.show()
    widgets.interact(channel,d=k2)
widgets.interact(genre,m=k1)

```

```python
p = widgets.Dropdown(
    options=video['channelTitle'].unique(),
    value=video['channelTitle'].unique()[0],
    description='channel:',
    disabled=False,
)
def channel(m):
    gen=data[data['channelTitle']==m]
    k2 = widgets.Dropdown(
        options=gen['title'].unique(),
        value=gen['title'].unique()[0],
        description='video:',
        disabled=False,
    )
    def vid(d):
      ht = gen[gen['title']==d]
      sns.lineplot(y = 'trending_date',x = 'view_count',data=ht)
      plt.title(d)
      plt.show()
    widgets.interact(vid,d=k2)
widgets.interact(channel,m=p)
```
